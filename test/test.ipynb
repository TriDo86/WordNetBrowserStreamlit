```python
def test_function(func, test_cases, expected_results):
    is_ok = True
    for i, (test, res) in enumerate(zip(test_cases, expected_results)):
        output = func(test)
        if output != res:
            print(f'Fail testcase ({i}):\noutput= {output}\nexpect= {res}')
            is_ok = False
    if is_ok:
        print(f'All test cases passed!')
```
